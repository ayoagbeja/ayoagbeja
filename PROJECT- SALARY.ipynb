```python
y = data['Salary']

x1 = data['YearsExperience']
```

```python
plt.scatter(x1, y)
plt.xlabel('YearsExperience')
plt.ylabel('Salary')

yhat = 8731.9410 *x1 +2.886e+04
fig = plt.plot(x1, yhat, lw= 3, c = 'red', label= 'regression line')

plt.title('Salary vs Years of Experience')
```
