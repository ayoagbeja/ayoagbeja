```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

The libraries used :
- Pandas to create and help tabulate the dataframe
- requests this enables HTTP requests to be sent via Python
- beautiful soup that makes it easy to scrape information from web pages


```python
url = 'https://www.imdb.com/list/ls560585364/'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
soup
```

```python
m_title = []
m_year = []
m_genre = []
metascore = []
m_synopsis = []
m_credit = []
```

```python
cast_data = soup.findAll('div', attrs= {'class': "lister-item-content"})
```

```python
for store in cast_data:
  name = store.h3.a.text
  m_title.append(name)

  year_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', ' ').replace(')', ' ')
  m_year.append(year_release)

  genre = store.p.find('span', class_ = 'genre')
  m_genre.append(genre)

  m_score = store.find( 'div',class_ ="inline-block ratings-metascore")

```
